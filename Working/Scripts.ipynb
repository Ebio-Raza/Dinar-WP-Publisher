### **Click Button**

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

def click_button(driver, selector, by="css", timeout=10):
    """
    Click a button or element on the page.

    Parameters:
        driver : selenium.webdriver
            The Selenium WebDriver instance.
        selector : str
            The CSS selector, XPath, or ID of the element.
        by : str
            How to locate the element: "css", "xpath", or "id".
        timeout : int
            Maximum wait time in seconds for the element to be clickable.
    """
    by_dict = {
        "css": By.CSS_SELECTOR,
        "xpath": By.XPATH,
        "id": By.ID
    }

    if by not in by_dict:
        raise ValueError("Invalid 'by' value. Use 'css', 'xpath', or 'id'.")

    element = WebDriverWait(driver, timeout).until(
        EC.element_to_be_clickable((by_dict[by], selector))
    )
    element.click()
```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

def close_popup_in_iframes(driver, button_selector="button.close", timeout=5):
    """
    Tries to find and click a close button inside any iframe.
    
    Parameters:
        driver : Selenium WebDriver
        button_selector : CSS selector of the close button
        timeout : max wait per iframe
    """
    iframes = driver.find_elements(By.TAG_NAME, "iframe")
    for index, iframe in enumerate(iframes):
        try:
            driver.switch_to.frame(iframe)
            
            close_button = WebDriverWait(driver, timeout).until(
                EC.element_to_be_clickable((By.CSS_SELECTOR, button_selector))
            )
            
            # Use JS click if normal click fails
            try:
                close_button.click()
            except:
                driver.execute_script("arguments[0].click();", close_button)
            
            print(f"Popup closed in iframe index: {index}")
            driver.switch_to.default_content()
            return True  # Successfully closed
        except:
            driver.switch_to.default_content()
            continue
    print("No popup found in any iframe")
    return False
```

### **Playwright**

```python
# For Jupyter notebooks
from playwright.async_api import async_playwright
import asyncio

async def main():
    async with async_playwright() as p:
        # Launch Chromium
        browser = await p.chromium.launch(headless=False)  # headless=True for headless mode
        
        # Create a new browser context
        context = await browser.new_context(
            viewport={"width": 1920, "height": 1080}  # similar to maximized
        )
        
        # Open a new page
        page = await context.new_page()
        
        # Navigate to URL
        await page.goto("https://www.dinarguru.com/")
        
        print("Browser is open. Close it manually when done.")
        # Note: Do not close immediately in Jupyter, let user interact
        # await browser.close()  # You can close manually when finished

# Run the async function in Jupyter
await main()
```
